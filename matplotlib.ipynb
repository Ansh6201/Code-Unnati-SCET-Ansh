# Sub-Plot

```python
import numpy as np
```

```python
a=np.array([36,46,56,66,76,86])
b=np.array([32,42,52,62,72,82])
c=np.array([35,45,55,65,75,85])
d=np.array([30,40,50,60,70,80])
plt.plot(a,b)
```

```python
a=np.array([36,42,56,60,76,86])
b=np.array([32,42,63,62,72,82])
c=np.array([37,45,55,69,79,85])
d=np.array([30,42,50,60,70,80])
plt.plot(a,b)
```

```python
# Subplot command: plt.subplot(no. of rows,no. of columns,position)
```

```python
#Plot 1
plt.subplot(2,2,1)
plt.plot(a,b)
```

```python
#Plot 2
plt.subplot(2,2,2)
plt.plot(b,c)
```

```python
#Plot 3
plt.subplot(2,2,3)
plt.plot(c,d)
```

```python
#Plot 4
plt.subplot(2,2,4)
plt.plot(d,a)
```

```python
#Plot 1
plt.subplot(2,2,1)
plt.plot(a,b)
#Plot 2
plt.subplot(2,2,2)
plt.plot(b,c)
#Plot 3
plt.subplot(2,2,3)
plt.plot(c,d)
#Plot 4
plt.subplot(2,2,4)
plt.plot(d,a)
```

```python
#Plot 1
plt.subplot(3,2,1)
plt.plot(a,b)
#Plot 2
plt.subplot(3,2,2)
plt.plot(b,c)
#Plot 3
plt.subplot(3,2,3)
plt.plot(c,d)
#Plot 4
plt.subplot(3,2,4)
plt.plot(d,a)
```

# Histogram

```python
data=np.random.normal(100,10,200)
```

```python
#Generating 200 random numbers
data
```

```python
len(data)
```

```python
#Plotting Histogram
plt.hist(data)
plt.show
```

```python
#Histogram is used to determine the frequency of a particular entry; y axis shows the frequency of the entry.
data=np.array([30,40,50,50,60,70,80])
plt.hist(data)
plt.show
```

```python
#bin size(bins) command is used to determine groups in histogram
plt.hist(data,bins=20)
plt.show
```

```python
data=np.random.normal(100,10,200)
plt.hist(data,bins=20)
plt.show
```

```python
data=np.array([65,72,86,80,76,96,92,83,65,85,90,92,84,77,85,86,76,77])
plt.hist(data,bins=10)
plt.show
```
